```python
years = [2006 + x for x in range(16)]
weights = [80, 83, 84, 85, 86, 82, 81, 79, 83, 80,
           82, 82, 83, 81, 80, 79]

plt.plot (years, weights, lw=3, linestyle="--")
plt.show()
```

```python
# dados categricos
x = ["C++", "C#", "Python", "Java", "Go"]
y = [20, 50, 140, 1, 45]

plt.bar(x, y, color='darkblue', align='edge', width=0.5, edgecolor='lightblue', lw=3)
plt.show()
```

```python
ages = np.random.normal(20, 1.5, 1000)
plt.hist(ages, bins=[ages.min(), 18, 19, 20, 21, 22, ages.max()], edgecolor='black')
plt.show()
```

```python
# cumulativo
plt.hist(ages, cumulative=True, edgecolor='black')
plt.show()
```

```python
# as categorias no se sobrepem
langs = ['Python', 'C++', 'Java', 'C#', 'Go']
votes = [50, 24, 14, 6, 17]
explodes = [0.2, 0, 0, 0.2, 0]

plt.pie(votes, labels=langs, explode=explodes,
        autopct='%.2f%%', pctdistance=0.8)
plt.show()
```

```python
heights = np.random.normal(172, 8, 300)
plt.boxplot(heights)
plt.show()
```

```python
years = [2014, 2015, 2016, 2017,
         2018, 2019, 2020, 2021]

income = [55, 56, 62, 61, 72, 72, 73, 75]

income_ticks = list(range(50, 81, 2))

plt.plot(years, income)
plt.title("Income of John (in USD)", fontsize=20) #fontname="" para mudar a fonte
plt.xlabel("Year")
plt.ylabel("Yearly income in USD")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])

plt.show()
```

```python
stock_a = [100, 102, 99, 101, 101, 100, 102]
stock_b = [90, 95, 102, 104, 105, 103, 109]
stock_c = [110, 115, 100, 105, 100, 98, 95]

plt.plot(stock_a, label="Company 1")
plt.plot(stock_b, label="Company 2")
plt.plot(stock_c, label = "Company 3")

plt.legend(loc="lower center")

plt.show()
```

```python
votes = [10, 2, 5, 16, 22]
people = ["A", "B", "C", "D", "E"]

style.use("ggplot")
plt.pie(votes, labels=None)
plt.legend(labels=people)
plt.show()
```

```python
x = np.arange(100)

fig, axs = plt.subplots(2, 2, figsize=(12, 8))


axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random Function")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Log function")

plt.show()
```
